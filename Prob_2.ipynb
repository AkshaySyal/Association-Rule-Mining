```python
url = "http://fimi.uantwerpen.be/data/kosarak.dat"
response = requests.get(url)
data = response.text
lines = data.split('\n')
```

```python
int_lines = []
for line in lines:
  curr_line = []
  for item in line.split():
    curr_line.append(int(item))
  int_lines.append(curr_line)
```

```python
max_id = 0
for line in int_lines:
  max_id = max(max_id, max(line))
```

```python
filename = "/content/drive/MyDrive/USML/HW1/PROB_2/cleanedSparseClickData.arff"
```

```python
def write_arff(filename, data, max_id):
  header = f"@relation cleanedSparseClickData\n"

  for i in range(1, max_id+1):
    header += f"@attribute {i} "
    header += "{0, 1}\n"

  header += "@data\n"

  data_section = ""
  for line in data:
    line = list(set(line)) # done for the purpose of data cleaning
    line.sort()
    data_section += "{"
    for i in range(len(line)-1):
      data_section += f"{line[i]-1} 1,"
    data_section += f"{line[-1]-1} 1"
    data_section += "}\n"

  data_section = data_section[:-1] # Remove the last newline character


  with open(filename, "w", encoding="utf-8") as file:
    file.write(header)
    file.write(data_section)
```

```python
write_arff(filename, int_lines, max_id)
```

```python
end_time = time.time()
print(f"Time taken: {end_time - start_time}")
```
